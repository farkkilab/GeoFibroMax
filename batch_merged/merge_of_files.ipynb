```python

genes = ['GeneA', 'GeneB', 'GeneC', 'GeneD', 'GeneE']
samples = ['Sample1', 'Sample2', 'Sample3']

# Randomized data for demonstration
np.random.seed(42)
gene_expression = np.random.rand(3, 5) * 10  # Simulating gene expression values (0-10)
tcell_abundance = np.random.rand(3, 5) * 100  # Simulating T-cell abundance values (0-100)

df_gene_expression = pd.DataFrame(gene_expression, columns=genes, index=samples)
df_tcell_abundance = pd.DataFrame(tcell_abundance, columns=genes, index=samples)

# Create a plot using scatter (for bubble chart effect)
plt.figure(figsize=(10, 6))

# Loop through each gene and create a scatter plot
for i, gene in enumerate(genes):
    # Plot gene expression vs. T-cell abundance
    plt.scatter(
        x=[gene] * len(samples), 
        y=samples, 
        s=df_tcell_abundance[gene] * 10,  # Bubble size (T-cell abundance scaled)
        c=df_gene_expression[gene],  # Bubble color (gene expression)
        cmap='viridis',  # Color map for expression levels
        alpha=0.7,  # Transparency for the bubbles
        label=gene
    )

# Customize the plot
plt.title('Bubble Chart: T-cell Abundance vs Gene Expression')
plt.xlabel('Genes')
plt.ylabel('Samples')
plt.colorbar(label='Gene Expression')  # Color bar for gene expression
plt.legend(title='Genes', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
